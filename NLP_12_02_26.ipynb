```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = ''' Dr. Smith lives in the U.S. and works at New York University.
He specializes in A.I. and has published over 3.14 papers every year .
He often travels to Washigton D.C. for conferences, workshops, etc.
Recently, he received an award from the A.I Socity.'''
```

```python
doc = nlp(text)
```

```python
type(doc)
```

```python
for s in doc.sents:
  print(s)
```

```python
text = "Hello! Are you coming?"
```

```python
doc = nlp(text)
```

```python
for token in doc:
  print(token,"_",token.pos_)
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
words = ["eating","eats","ate","adjustable","rafting","ability","meeting","better","beutiful"]
```

```python
for word in words:
  print(word," ",ps.stem(word))
```
